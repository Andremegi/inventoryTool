# Caducidad

```python
interes_caducidad=['Código', 'Descripción', 'Familia','S.Actual','S.Minimo', 'S.Maximo','Rotación','Caducidad']
```

```python
#Quitamos los valores desconocidos de Caducidad y los rellenamos con unknown
df_caducidad = df[interes_caducidad]
df_caducidad['Caducidad'].fillna('Unknown', inplace=True)
```

```python
#Quitamos los valores desconocidos pues no nos interesan para este fin 
df_caducidad_act = df_caducidad[df_caducidad['Caducidad']!='Unknown']
```

```python
#Cambiamos formato para que sea de fecha y creamos una copia con lo nuevo
df_caducidad_act['Caducidad_act'] = pd.to_datetime(df_caducidad_act['Caducidad'], format="%d/%m/%y", yearfirst=False)
```

```python
df_caducidad_act
```

## Productos ya caducados 
- Vamos a explorar productos que ya esten caducados , a partir de la fecha que sea menor que la de hoy 
- Exploramos que el stock actual S.Actual > 0 ya que esos son los productos que tenemos en la farmacia

```python

df_productos_caducados = df_caducidad_act[(df_caducidad_act['Caducidad_act'] < pd.Timestamp.today()) & (df['S.Actual']>0)]
df_productos_caducados
```

## Caducidad de 2026
1. Obtenemos el año actual 
2. Vemos productos que tenemos en Stock , S.Actual > 0
3. Vemos productos cuyo año es el que queremos (año actual)

```python
año_actual = pd.Timestamp.today().year
df_productos_van_caducar = df_caducidad_act[(df_caducidad_act['Caducidad_act'].dt.year == año_actual) & (df['S.Actual']>0)]
df_productos_van_caducar.sort_values('Caducidad_act', ascending = True, inplace= True)
```

# STOCK  
La rotación debe ser mayor que 0 pues esto indica que el producto es de interes y se vende , para el aviso de stock necesitamos 
1. Que la Rotacion sea mayor que 0 
2. Que los productos sean de las familias de interes

```python
familias_interes=['DIETETICA Y HERBORISTERIA','SOLARES','COSMÉTICA','CORPORAL']
interes_caducidad
```

```python
df_interes_stock = df[df['Familia'].isin(familias_interes) & df['Rotación']>0 ][interes_caducidad]
df_interes_stock
```

```python
df_interes_stock_stock_caducidad = df_interes_stock[df_interes_stock['Caducidad'].isna()]
df_interes_stock_stock_caducidad
```

```python
df_familias_interess_sin_rotacion = df[df['Familia'].isin(familias_interes) & (df['S.Actual']!=0) & (df['Rotación']==0 | df['Rotación'].isna()) ][interes_caducidad]
```

```python
with pd.ExcelWriter('reportes1.xlsx', engine='openpyxl') as writer:
    df_productos_caducados.to_excel(writer, sheet_name='Caducados', index=False)
    df_productos_van_caducar.to_excel(writer, sheet_name='Caducan 26', index=False)
    df_interes_stock_stock_caducidad.to_excel(writer, sheet_name='stock sin caducidad', index=False)
    df_familias_interess_sin_rotacion.to_excel(writer, sheet_name='stock sin rotacion', index=False)
```
